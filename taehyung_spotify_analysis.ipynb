```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("taehyung_spotify_streams.csv")
df.head(10)
```

# مرتب‌سازی بر اساس تعداد استریم

```python
df_sorted = df.sort_values(by='streams', ascending=False)
```

# تنظیم فونت فارسی

```python
plt.rcParams['font.family'] = 'DejaVu Sans'
```

# نمودار میله‌ای استریم‌ها

```python
plt.figure(figsize=(12,6))
sns.barplot(x='streams', y='track_name', data=df_sorted, color='purple')
plt.title(' GLOBAL STREAMS Taehyung in Spotify', fontsize=16)
plt.xlabel('Streams', fontsize=12)
plt.ylabel('Songs Name', fontsize=12)
plt.tight_layout()
plt.show()
```
